### point 클래스 생성
- 2차원 좌표평면 각 점(x, y)

### 지원 메소드
- 두 점의 덧셈, 뺄셈 ex). (1, 2) + (3, 4) = (4, 6)
- 한 점과 숫자의 곱셈 ex). (1, 2) * 3 = (3, 6)
- 원 점(0, 0)에서 부터 그 점까지의 거리
- x, y 값 가져오기
- 출력 하기

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    # 커스텀해서 직접 출력할 수 있지만
    # 기존 print를 override해서 print로 아래와 같은 기능을 구현 할 수 있다
    def print_pt(self):
        print('({}, {})'.format(self.x, self.y))
        
    # 입력 받은 값을 문자열 화 하라는 내장 메소드
    def __str__(self):
        return '({}, {})'.format(self.x, self.y)
    
    def add(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    
    # 더하기 연산자
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    
    # 빼기 연산자
    def __sub__(self, pt):
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point(new_x, new_y)
    
    def multiply(self, factor):
        return Point(self.x * factor, self.y * factor)
    
    # 곱하기 연산자
    def __mul__(self, factor):
        return Point(self.x * factor, self.y * factor)
    
    def length(self):
        return self.x ** 2 + self.y ** 2
    
    # len메소드
    def __len__(self):
        return self.x ** 2 + self.y ** 2
    
    def get_x(self):
        return self.x
    
    def get_y(self):
        return self.y
        
    # indexing 처리
    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        else:
            return -1
    
p1 = Point(3, 4)
p2 = Point(2, 7)

# p3 = p1.add(p2)
p3 = p1 + p2

# p4 = p1.sub(p2)
p4 = p1 - p2

# p5 = p1.multiply(3)
p5 = p1 * 3

# print(p1.length())
print(len(p1))

# print(p1.get_x())
print(p1[0])

# print(p1.get_y())
print(p1[1])

print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
```
