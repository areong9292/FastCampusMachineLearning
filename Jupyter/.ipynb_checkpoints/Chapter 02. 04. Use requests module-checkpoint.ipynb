## Get 방식
- 페이지 주소(Url)에 데이터가 들어있다
- 데이터 공유가 가능

## Post 방식
- Form에 데이터를 전송하므로 주소에 데이터가 노출되지 않는다

### requests 모듈
- http reauest/response를 위한 모듈
- hppt method를 메소드 명으로 사용하여 request 요청

```python
import requests
```

### get 요청하기
- http get 요청하기
- query parameter 이용하여 데이터 전달하기

```python
# 이 주소의 정보를  가져와라
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)

resp
```

```python
# 데이터 전체
# 페이지 소스보기로 본 내용과 같다
resp.text
```

### post 요청하기
- http post 요청하기
- post data 이용하여 데이터 전달하기

```python
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id': 'macmath22',
    'pwd': 'Test1357!'
}

# post는 form을 보내야한다
# dic 형태의 data를 전송한다
resp = requests.post(url, data=data)
resp.text

# 이 내용은 로그인 시도로 pwd에 다른 값을 넣으면 실패한다
```

```python
# 크롤링할 때 헤더를 쓰면 된다는데
# 일단 헤더를 이렇게 넣을 수 있다만 알려줌
url = 'https://news.v.daum.net/v/20190728165812603'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
}
resp = requests.get(url, headers=headers)

resp.text
```

### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)

# 응답 코드
print(resp.status_code)

if resp.status_code==200:
    print("success")
else:
    print("error")
```
